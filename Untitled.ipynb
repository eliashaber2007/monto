```python
import numpy as np
import matplotlib.pyplot as plt

age = np.array([1, 20, 40, 60, 80, 100, 120])

dominant    = np.array([1.26, 22.29, 40.09, 56.15, 63.49, 71.69, 81.08])
codominant  = np.array([1.27, 16.02, 29.42, 31.61, 35.61, 35.69, 35.93])
subdominant = np.array([1.29, 22.14, 35.69, 49.23, 56.88, 60.43, 63.74])

plt.figure(figsize=(8, 5))
plt.plot(age, dominant,    marker='o', label='Dominant')
plt.plot(age, codominant,  marker='s', label='Codominant')
plt.plot(age, subdominant, marker='^', label='Subdominant')

plt.xlabel('Age (years)')
plt.ylabel('Circumference (cm)')
plt.title('Trunk circumference vs age — three pine categories')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
