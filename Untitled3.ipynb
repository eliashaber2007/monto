```python
import numpy as np
import matplotlib.pyplot as plt

Ro = np.array([1, 2, 5, 10, 15, 20, 50, 100, 200, 500, 1000, 2500, 5000, 6000])
V = np.array([0.0189, 0.0367, 0.0899, 0.1602, 0.2442, 0.3445, 0.4474, 0.4647, 0.4714, 0.4750, 0.4750, 0.4760, 0.4762, 0.4763])
I = V / Ro
P = V * I

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

ax1.plot(V, I, 'bo-')
ax1.set_xlabel('V (V)')
ax1.set_ylabel('I (A)')
ax1.set_title('I–V Curve')
ax1.grid(True)

ax2.plot(V, P, 'ro-')
ax2.set_xlabel('V (V)')
ax2.set_ylabel('P (W)')
ax2.set_title('Power Curve')
ax2.grid(True)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Dataset 1 - 17cm
Ro1 = np.array([1, 2, 5, 10, 15, 20, 50, 100, 200, 500, 1000, 2500, 5000, 6000])
V1 = np.array([0.0189, 0.0367, 0.0899, 0.1602, 0.2442, 0.3445, 0.4474, 0.4647, 0.4714, 0.4750, 0.4750, 0.4760, 0.4762, 0.4763])
I1 = V1 / Ro1
P1 = V1 * I1

# Dataset 2 - 10cm
Ro2 = np.array([1, 2, 5, 10, 15, 20, 50, 100, 200, 500, 1000, 2500, 5000, 6000])
V2 = np.array([0.0443, 0.0862, 0.2104, 0.3642, 0.4422, 0.4689, 0.4931, 0.499, 0.5015, 0.503, 0.5033, 0.5035, 0.5033, 0.5032])
I2 = V2 / Ro2
P2 = V2 * I2

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

ax1.plot(V1, I1, 'bo-', label='h=17cm')
ax1.plot(V2, I2, 'ro-', label='h=10cm')
ax1.set_xlabel('V (V)')
ax1.set_ylabel('I (A)')
ax1.set_title('I–V Curve')
ax1.legend()
ax1.grid(True)

ax2.plot(V1, P1, 'bo-', label='h=17cm')
ax2.plot(V2, P2, 'ro-', label='h=10cm')
ax2.set_xlabel('V (V)')
ax2.set_ylabel('P (W)')
ax2.set_title('Power Curve')
ax2.legend()
ax2.grid(True)

plt.tight_layout()
plt.show()
```
